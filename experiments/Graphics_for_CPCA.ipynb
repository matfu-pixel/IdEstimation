```python
import numpy as np
import skdim 
from matplotlib import pyplot as plt
import estimators
```

```python
benchmark = skdim.datasets.BenchmarkManifolds(random_state=0)
dict_data = benchmark.generate()
```

### Проверка сходимости CPCA на множестве M1_Sphere без шума

```python
benchmark = skdim.datasets.BenchmarkManifolds(random_state=0)
dict_data = benchmark.generate()
x = []
y = []
for k in range(10, 200, 15):
    x.append(k)
    cpca = estimators.cPCA()
    cpca.fit(dict_data['M1_Sphere'], n_neighbors=k, noise=True)
    y.append(cpca.dimension_)
plt.plot(x, y)
```

### Проверка сходимости CPCA на множестве Mbeta	без шума
